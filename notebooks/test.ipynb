```python
import pandas as pd
import numpy as np
from pathlib import Path

raw_dir = Path(r'C:\Users\santa\Desktop\ETL_cositas\workshop_03\data\raw')

dfs = {y: pd.read_csv(raw_dir / f'{y}.csv') for y in [2015, 2016, 2017, 2018, 2019]}

# Renombrar
RENAME_MAP = {
    2015: {'Country':'country','Region':'region','Happiness Rank':'rank','Happiness Score':'score','Economy (GDP per Capita)':'gdp','Family':'family','Health (Life Expectancy)':'health','Freedom':'freedom','Trust (Government Corruption)':'trust','Generosity':'generosity'},
    2016: {'Country':'country','Region':'region','Happiness Rank':'rank','Happiness Score':'score','Economy (GDP per Capita)':'gdp','Family':'family','Health (Life Expectancy)':'health','Freedom':'freedom','Trust (Government Corruption)':'trust','Generosity':'generosity'},
    2017: {'Country':'country','Happiness.Rank':'rank','Happiness.Score':'score','Economy..GDP.per.Capita.':'gdp','Family':'family','Health..Life.Expectancy.':'health','Freedom':'freedom','Trust..Government.Corruption.':'trust','Generosity':'generosity'},
    2018: {'Country or region':'country','Overall rank':'rank','Score':'score','GDP per capita':'gdp','Social support':'family','Healthy life expectancy':'health','Freedom to make life choices':'freedom','Perceptions of corruption':'trust','Generosity':'generosity'},
    2019: {'Country or region':'country','Overall rank':'rank','Score':'score','GDP per capita':'gdp','Social support':'family','Healthy life expectancy':'health','Freedom to make life choices':'freedom','Perceptions of corruption':'trust','Generosity':'generosity'},
}
FINAL_COLS = ['year','country','region','rank','score','gdp','family','health','freedom','generosity','trust']

for year, df in dfs.items():
    dfs[year] = df.rename(columns=RENAME_MAP[year])
    dfs[year]['year'] = year
    if 'region' not in dfs[year].columns:
        dfs[year]['region'] = np.nan
    dfs[year] = dfs[year][[c for c in FINAL_COLS if c in dfs[year].columns]]

# Fill regions
region_map = dfs[2015].set_index('country')['region'].to_dict()
for year in [2017, 2018, 2019]:
    dfs[year]['region'] = dfs[year]['country'].map(region_map)

# Concatenar
df = pd.concat(dfs.values(), ignore_index=True)

# Ver nulos
print("Nulos por columna:")
print(df.isnull().sum())
print("\nFilas con nulos:")
print(df[df.isnull().any(axis=1)][['year', 'country', 'region']].to_string())
```
